## "$\color{red}{\text{Make the Receiver-Function for One station and Compare it with Paper in ObsPy}}$"


## To have the Reciever_function for one station, at first, we need to input the three components of the earthquake in(E, N, Z) directions, which has a good quality SNR ratio, and then rotate them in radial and tangential directions. I chose the teleseismic events on the IRIS website, which are bigger than 6 in magnitude. We selected in a range of (10km <depth< 20 km), and the ones between 30 to 95 degrees as distance to the station. For the rotation process, we must carefully separate specific parts of the traces to ensure we have the first "P"  arrivals. So, in most of them, I selected a time window of around at least 30(s) before P arrival and about 10(s) after P arrival, I tried to remove surface waves in the time window as well. About the filter,a bandpass filter (0.03 to 0.3 to 1 Hz) was applied after the rotation code. Source and trace are single traces, and then we must produce a single deconvolved  trace. Water-level selected by trial and error between (0.001,0.01,0.1,0.5) values. Then I calculated the mean of the receiver function of the events recorded by that station to improve the quality of the result, and at the last  step,I compared the mean figure with Schiffer's receiver function for the TULEG station.
